```python
import pandas as pd
```

```python
d1 = pd.read_csv('BookScrap.csv')
```

```python
d1.head()
```

```python
d2 = pd.read_csv('data.csv')
```

```python
d2.head()
```

```python
df = pd.DataFrame()
```

```python
df['Title'] = d1['title']
df['Type'] = d2['Type']
df['Rating'] = d1['rating']
df['Price'] = d1['price']
df['Price exc. Tax'] = d2['Price Excl. Tax']
df['Price inc. Tax'] = d2['Price Inc. Tax']
df['Tax'] = d2['Tax']
df['Quantity'] = d2['Quantity']
df['Book-Link'] = d1['book']
df['Image-Link'] = d1['img']
```
