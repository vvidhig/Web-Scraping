```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
data = []
for i in tqdm(range(1,51)):
    link = "https://books.toscrape.com/catalogue/page-" + str(i) +".html"
    res = requests.get(link)
    soup = BeautifulSoup(res.text, 'html.parser')
    for div in soup.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3'):
        img = 'https://books.toscrape.com/' + div.find('img', class_='thumbnail').get('src')
        book = 'https://books.toscrape.com/catalogue/' + div.find('a').get('href')
        title = div.find_all('a')[-1].get('title')
        rating = div.find('p').get('class')[-1]
        price = div.find('p', class_='price_color').text[2:]
        availability = div.find('p', class_='instock availability').text.strip()
        data.append([title, book, img, rating, price, availability])
```

```python
df = pd.DataFrame(data, columns=['title', 'book', 'img', 'rating', 'price', 'availability'])
```
